```python
#create some dummy variable since "purpose" column is in categorical tpye
final_data = pd.get_dummies(loans,columns=["purpose"],drop_first=False)
```

```python
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into test and train data sets
X = final_data.drop("not.fully.paid",axis=1)
y = final_data["not.fully.paid"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
#import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
#lets predict out test data
prediction = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
#we will compare the above decision tree classifier with random forest classifeir
#import randomforestclassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=700)
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

# random forest is much better on base in all the category(precision,recall, f1_score)
